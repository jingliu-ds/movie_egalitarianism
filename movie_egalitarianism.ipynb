```r
insertRow <- function(existingDF, newrow, r) {
  existingDF[seq(r+1,nrow(existingDF)+1),] <- existingDF[seq(r,nrow(existingDF)),]
  existingDF[r,] <- newrow
  existingDF
}
```

```r
attn <- insertRow(att, att[9,], 1)
attn2 <- attn[-10,] 
attn2
```
